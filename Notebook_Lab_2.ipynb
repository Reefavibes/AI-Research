# Normal Distribution

```python
# Define the parameters for the normal distribution
mean = 0  # Mean (mu) of the distribution
std_dev = 0.1  # Standard deviation (sigma) of the distribution

# Generate 1000 data points following a normal distribution with the specified mean and standard deviation
sample_size = 1000
data_points = np.random.normal(mean, std_dev, sample_size)

# Plot the histogram of the data points
bins_number = 30  # Number of bins for the histogram
hist_count, x, ignored = plt.hist(data_points, bins_number, density=True)

# Plot the probability density function of the normal distribution
normal_dist_curve = 1 / (std_dev * np.sqrt(2 * np.pi)) * np.exp(- (x - mean) ** 2 / (2 * std_dev ** 2))
plt.plot(x, normal_dist_curve, linewidth=2, color='r')

# Set the title and labels for the plot
plt.title('Normal Distribution Visualization')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Display the plot
plt.show()
```

```python
# Function to calculate the mean of a dataset
def mean(x):
    return np.sum(x) / x.shape[0]
```

```python
# Function to calculate the variance of a dataset
def var(x):
    return mean((x - mean(x))**2)
```

```python
# Function to calculate the standard deviation of a dataset
def std(x):
    return np.sqrt(var(x))
```

```python
# Function to calculate the median of a dataset
def median(x):
    n = len(x)
    sorted_x = np.sort(x)
    mid = n // 2
    if n % 2 == 0:  # If the number of data points is even
        return (sorted_x[mid - 1] + sorted_x[mid]) / 2
    else:  # If the number of data points is odd
        return sorted_x[mid]
```

```python
# Function to calculate the mode of a dataset
def mode(x):
    values, counts = np.unique(x, return_counts=True)
    max_count_index = np.argmax(counts)
    return values[max_count_index]
```

```python
# Function to calculate the range of a dataset
def range(x):
    return np.max(x) - np.min(x)
```

```python
# Example dataset
data = np.array([1, 2, 3, 4, 5, 5, 6])

# Testing the functions by calculating statistical measures of the dataset
mean_value = mean(data)
var_value = var(data)
std_value = std(data)
median_value = median(data)
mode_value = mode(data)
range_value = range(data)

# Printing the calculated values
mean_value, var_value, std_value, median_value, mode_value, range_value
```

# Pi from Random Numbers

```python
# Set the number of points to be generated
N = 10000
```

```python
# Generate an array of 'N' points with 2 coordinates. The points are uniformly distributed
# over the square [-1,1] x [-1,1], as 'np.random.random' generates random floats in the [0.0, 1.0) range
points = -1 + 2*np.random.random((N,2))
```

```python
# Plot the points. The ':' is used to select all elements in the respective dimension,
# so 'points[:,0]' is the x-coordinate and 'points[:,1]' is the y-coordinate.
plt.plot(points[:,0], points[:,1], '.')

# Set the aspect of the plot to be equal, ensuring that circles appear as circles,
# squares as squares, etc., in the plot, instead of being stretched.
plt.gca().set_aspect(1)
```

```python
# Define boolean arrays for points inside and outside the unit circle
inside_circle = points[:,0]**2 + points[:,1]**2 <= 1
outside_circle = points[:,0]**2 + points[:,1]**2 > 1

# Plot points inside the circle in green
plt.plot(points[inside_circle,0], points[inside_circle,1], 'g.')

# Plot points outside the circle in red
plt.plot(points[outside_circle,0], points[outside_circle,1], 'r.')

# Set the aspect ratio of the plot to be equal, ensuring accurate representation of the circle
plt.gca().set_aspect(1)
```

```python
# Sum the boolean arrays to count how many points are inside and outside the unit circle
np.sum(inside_circle), np.sum(outside_circle)
```

```python
# Since the points are distributed in a 2x2 square, the total area is 4
total_area = 4
```

```python
# Calculate the fraction of points that lie inside the unit circle
fraction_inside = np.sum(inside_circle) / N
```

```python
# Multiply the fraction of points inside by the total area to approximate the area of the circle,
# which is pi when the radius is 1. Therefore, this operation approximates pi.
fraction_inside * total_area
```

# Complete Code for Estimating  using Monte Carlo Simulation

```python
# Number of random points to generate
num_points = 10000

# Generating random points within the range [-1, 1] for both x and y coordinates
x_points = np.random.uniform(-1, 1, num_points)
y_points = np.random.uniform(-1, 1, num_points)

# Calculating the number of points inside the quarter circle with radius 1
# A point is inside the circle if the distance from the origin is less than or equal to 1
points_inside = np.sqrt(x_points**2 + y_points**2) <= 1
num_inside = np.sum(points_inside)

# Estimating  by multiplying the ratio of points inside the quarter circle by 4
pi_estimate = 4 * num_inside / num_points

# Plotting the points and the quarter circle
plt.figure(figsize=(6, 6))  # Set the figure size to be square
# Plot points inside the quarter circle in green
plt.scatter(x_points[points_inside], y_points[points_inside], color='green', label='Inside')
# Plot points outside the quarter circle in red
plt.scatter(x_points[~points_inside], y_points[~points_inside], color='red', label='Outside')
# Create a circle object to represent the quarter circle
circle = plt.Circle((0, 0), 1, color='blue', fill=False)
# Add the circle to the plot
plt.gca().add_artist(circle)

# Add a title and labels to the plot, and set the aspect ratio to be equal
plt.title('Estimating  using Monte Carlo Simulation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axis('equal')  # Ensures x and y have the same scale
plt.show()  # Display the plot

# Display the estimate of 
pi_estimate
```

# e from Random Numbers

```python
# Generate a 2D array 'X' with 1,000,000 rows and 10 columns, filled with random numbers between 0 and 1
X = np.random.random((1000000, 10))
```

```python
# Compute the cumulative sum 'Y' of 'X' along axis 1 (summing across columns)
Y = np.cumsum(X, 1)
```

```python
# Find the index (position) along axis 1 where the cumulative sum 'Y' first exceeds 1
# The result is an array 'Z' with the same number of rows as 'Y'
Z = np.argmax(Y > 1, 1) + 1
# Adding 1 to the result because Python uses zero-based indexing, and we want one-based indexing
```

```python
# Compute the mean of the indices in array 'Z'
# This average represents an experimental estimate of the constant 'e'
np.mean(Z)
```

```python
# Alternatively, you can calculate the estimated 'e' directly in one line
np.mean(np.argmax(np.cumsum(np.random.random((10000000,10)),1) > 1,1) + 1)
```
